```python
import numpy as SpaceInInput
import pandas as pd
import lime
import lime.lime_tabular
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
file_path='/content/HousingData.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
#missing value treatment
df.isnull().sum()
```

```python
df=df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
#Declare feature vectr and target variable
X = df[['LSTAT','RM','NOX', 'PTRATIO', 'DIS','AGE']]
Y=df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
model = RandomForestRegressor(max_depth=6, random_state=0, n_estimators=10
                              )
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import lime.tabular
```

```python
explainer = lime.lime_tabular.LimeTabularExplainer(X_train.values, feature_names=X_train.columns.values.tolist(),class_names=['MEDV'],
                                                  verbose=True, mode='regression', random_state=1)
```

```python
j=5
exp = explainer.explain_instance(X_test.values[j], model.predict, num_features=6)
```
